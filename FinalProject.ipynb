```python
import numpy as np
import pandas as pd
```

```python
airbnb_data = pd.read_csv('dataset/airbnb.csv')
airbnb_data
```

```python
airbnb_data.info()
```

```python
airbnb_data.describe(include = "all")
```

```python
airbnb_data.dropna(subset=['name', 'host_name'], inplace=True)
```

```python
values = airbnb_data["last_review"].value_counts().index.tolist()
counts = airbnb_data["last_review"].value_counts().tolist()
val_counts = dict(zip(values, counts))
val_counts
```

```python
airbnb_data["last_review"] = airbnb_data["last_review"].fillna('2019-06-23')
```

```python
airbnb_data["reviews_per_month"] = airbnb_data["reviews_per_month"].fillna('1.373229')
```

```python
airbnb_data
```

```python
airbnb_data["last_review"] = pd.to_datetime(airbnb_data["last_review"],format= "%Y-%m-%d")
```

```python
airbnb_data["reviews_per_month"] = pd.to_numeric(airbnb_data["last_review"], downcast="float")
```

```python
#Make Price Category
airbnb_data["price_category"] = np.where(airbnb_data["price"] > 106, "Expensive", (np.where(airbnb_data["price"] > 69, "Medium", "Cheap")))
airbnb_data
```

```python
values = airbnb_data["price_category"].value_counts().index.tolist()
counts = airbnb_data["price_category"].value_counts().tolist()
val_counts = dict(zip(values, counts))
val_counts
```

```python
#Filter the data with last review > 2019
airbnb_data = airbnb_data.loc[airbnb_data["last_review"] >= "2019-01-01"]
```
